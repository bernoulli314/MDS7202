```python
import os
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chi2_contingency
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import FunctionTransformer, MinMaxScaler
from sklearn.pipeline import Pipeline
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import shutil
from sklearn.ensemble import IsolationForest
```

```python
class Profiler:
    def __init__(self, data):
        if not isinstance(data, pd.DataFrame):
            raise ValueError("Se espera un objeto de tipo pandas DataFrame como entrada")

        now = datetime.datetime.now()
        folder_name = 'EDA_' + now.strftime("%d-%m-%Y")
        if not os.path.exists(folder_name):
            os.mkdir(folder_name)
            os.mkdir(os.path.join(folder_name, "clean_data"))
        
        plots_folder = os.path.join(folder_name, 'plots')
        if not os.path.exists(plots_folder):
            os.mkdir(plots_folder)

        self.data = data
        self.output_folder = folder_name

    def get_data(self):
        return self.data

    def get_output_folder(self):
        return self.output_folder

    def summarize(self, columns_of_interest=None, threshold_outliers=3):
        if columns_of_interest is None:
            columns_of_interest = self.data.columns

        summary_file = os.path.join(self.output_folder, 'summary.txt')

        with open(summary_file, 'w') as f:
            f.write("Resumen del Dataset:\n")

            for column in columns_of_interest:
                if column in self.data.columns:
                    f.write(f"Variable: {column}\n")
                    data_column = self.data[column]

                    # Reportar el tipo de variable
                    f.write(f"Tipo de Variable: {data_column.dtype}\n")

                    # Reportar el n√∫mero y porcentaje de valores √∫nicos
                    unique_count = data_column.nunique()
                    total_count = len(data_column)
                    unique_percent = (unique_count / total_count) * 100
                    f.write(f"N√∫mero de Valores √önicos: {unique_count} ({unique_percent:.2f}%)\n")

                    # Reportar el n√∫mero y porcentaje de valores nulos
                    null_count = data_column.isnull().sum()
                    null_percent = (null_count / total_count) * 100
                    f.write(f"N√∫mero de Valores Nulos: {null_count} ({null_percent:.2f}%)\n")

                    # Si la variable es num√©rica
                    if pd.api.types.is_numeric_dtype(data_column):
                        f.write("Estad√≠sticas Descriptivas:\n")
                        f.write(f"Valor M√≠nimo: {data_column.min()}\n")
                        f.write(f"Percentil 25: {data_column.quantile(0.25)}\n")
                        f.write(f"Media: {data_column.mean()}\n")
                        f.write(f"Percentil 50 (Mediana): {data_column.median()}\n")
                        f.write(f"Percentil 75: {data_column.quantile(0.75)}\n")
                        f.write(f"Valor M√°ximo: {data_column.max()}\n")

                        # Reportar el n√∫mero y porcentaje de valores cero
                        zero_count = (data_column == 0).sum()
                        zero_percent = (zero_count / total_count) * 100
                        f.write(f"N√∫mero de Valores Cero: {zero_count} ({zero_percent:.2f}%)\n")

                        # Reportar el n√∫mero y porcentaje de valores negativos
                        negative_count = (data_column < 0).sum()
                        negative_percent = (negative_count / total_count) * 100
                        f.write(f"N√∫mero de Valores Negativos: {negative_count} ({negative_percent:.2f}%)\n")

                        # Reportar el n√∫mero y porcentaje de valores at√≠picos (outliers)
                        iqr = data_column.quantile(0.75) - data_column.quantile(0.25)
                        lower_bound = data_column.quantile(0.25) - threshold_outliers * iqr
                        upper_bound = data_column.quantile(0.75) + threshold_outliers * iqr
                        outliers_count = ((data_column < lower_bound) | (data_column > upper_bound)).sum()
                        outliers_percent = (outliers_count / total_count) * 100
                        f.write(f"N√∫mero de Outliers: {outliers_count} ({outliers_percent:.2f}%)\n")

                    f.write("\n")
                else:
                    f.write(f"Variable '{column}' no encontrada en el DataFrame.\n\n")
            
        with open(summary_file, 'r') as texto:
            read_file = texto.read()
        
        print(read_file)
    
    def plot_vars(self, columns_of_interest=None, top_n_categories=10):
        if columns_of_interest is None:
            columns_of_interest = self.data.columns

        plots_folder = os.path.join(self.output_folder, 'plots')

        for column in columns_of_interest:
            if column in self.data.columns:
                data_column = self.data[column]
                plt.figure(figsize=(10, 8))

                if pd.api.types.is_numeric_dtype(data_column):
                    # Genere un gr√°fico de distribuci√≥n de densidad para variables num√©ricas
                    sns.histplot(data_column, kde=True, color='b')
                    plt.title(f'Distribuci√≥n de Densidad de {column}')
                    plt.xlabel(column)
                    plt.ylabel('Densidad')
                    plt.show()
                    file_path = os.path.join(plots_folder, f'{column}.pdf').replace("\\", "/")
                    plt.savefig(file_path)
                    plt.close()

                    # Grafique la correlaci√≥n entre las variables num√©ricas
                    if len(columns_of_interest) > 1:
                        for col2 in columns_of_interest:
                            if col2 != column and pd.api.types.is_numeric_dtype(self.data[col2]):
                                sns.scatterplot(x=column, y=col2, data=self.data, alpha=.2)
                                plt.title(f'Correlaci√≥n entre {column} y {col2}')
                                plt.xlabel(column)
                                plt.ylabel(col2)
                                plt.show()
                                file_path = os.path.join(plots_folder, f'{column}_vs_{col2}.pdf').replace("\\", "/")
                                plt.savefig(file_path)
                                plt.close()
                elif (isinstance(data_column.dtype, pd.CategoricalDtype) or not pd.api.types.is_numeric_dtype(data_column)):
                    # Genere un histograma de las top N categor√≠as
                    top_categories = data_column.value_counts()[:top_n_categories]
                    sns.barplot(x=top_categories.index, y=top_categories.values)
                    plt.title(f'Histograma de las Top {top_n_categories} Categor√≠as de {column}')
                    plt.xlabel(column)
                    plt.ylabel('Frecuencia')
                    plt.xticks(rotation=45)
                    plt.show()
                    file_path = os.path.join(plots_folder, f'{column}.pdf').replace("\\", "/")
                    plt.savefig(file_path)
                    plt.close()

                    # Grafique el coeficiente V de Cramer entre las variables categ√≥ricas
                    if len(columns_of_interest) > 1:
                        for col2 in columns_of_interest:
                            if col2 != column and (isinstance(self.data[col2].dtype, pd.CategoricalDtype) or not pd.api.types.is_numeric_dtype(self.data[col2])):
                                confusion_matrix = pd.crosstab(data_column, self.data[col2])
                                chi2, _, _, _ = chi2_contingency(confusion_matrix)
                                n = self.data.shape[0]
                                v_cramer = np.sqrt(chi2 / (n * (min(confusion_matrix.shape) - 1)))
                                sns.heatmap(np.array([[v_cramer]]), annot=True, cmap='coolwarm', fmt=".2f", cbar=False)
                                plt.title(f'Coeficiente V de Cramer entre {column} y {col2}')
                                plt.show()
                                file_path = os.path.join(plots_folder, f'{column}_vs_{col2}_cramer.pdf').replace("\\", "/")
                                plt.savefig(file_path)
                                plt.close()

    def clean_data(self, columns_of_interest=None, drop_duplicates=True, impute_technique=None, separator=None):
        
        if columns_of_interest is None:
            columns_of_interest = self.data.columns

        for column in columns_of_interest:
            if column not in self.data.columns:
                raise ValueError(f"Columna '{column}' no encontrada en el DataFrame")

        clean_data_folder = os.path.join(self.output_folder, 'clean_data')
        if not os.path.exists(clean_data_folder):
            os.mkdir(clean_data_folder)

        # Crear una copia del DataFrame para el procesamiento
        cleaned_data = self.data.copy()
        c_data_interest = cleaned_data[columns_of_interest].copy()

        dropped_columns = []
        added_columns = []

        if separator != None:
            # Separar columnas no at√≥micas
            for column in columns_of_interest:
                if not pd.api.types.is_numeric_dtype(c_data_interest[column]):
                    c_data_interest[column] = c_data_interest[column].astype(str).replace(str(separator), ':', regex=True)
                    if c_data_interest[column].astype(str).str.contains(':').any():
                        new_columns = c_data_interest[column].str.split(':', expand=True)
                        new_columns.columns = [f"{column}_{i}" for i in range(new_columns.shape[1])]
                        added_columns.extend(new_columns)
                        c_data_interest = pd.concat([c_data_interest, new_columns], axis=1)
                        c_data_interest.drop(column, axis=1, inplace=True)
                        dropped_columns.append(column)
            columns_of_interest = c_data_interest.columns
            print(columns_of_interest)
        else:
            dropped_columns = columns_of_interest
            print("No se especific√≥ un separador. No se realizar√° la separaci√≥n de columnas.")

        
        # cambiar nan strings a nan de pandas
        tipos_nan = ['nan', 'NaN', 'NAN', 'Nan']
        for tipo in tipos_nan:
            c_data_interest.replace(tipo, np.nan, inplace=True)

        # Tratar valores nulos
        if impute_technique == 'drop':
            c_data_interest.dropna(subset=columns_of_interest, inplace=True)
        elif impute_technique == 'impute_mean':
            for column in columns_of_interest:
                c_data_interest[column] = pd.to_numeric(c_data_interest[column], errors='coerce')
                c_data_interest[column].fillna(c_data_interest[column].mean(), inplace=True)
        elif impute_technique == 'impute_zeros':
            for column in columns_of_interest:
                c_data_interest[column] = pd.to_numeric(c_data_interest[column], errors='coerce')
                c_data_interest[column].fillna(0, inplace=True)
        
        # quitar las columnas droppeadas del dataframe original
        if len(dropped_columns) > 0:
            cleaned_data.drop(dropped_columns, axis=1, inplace=True)

        if len(cleaned_data.columns) > 0:
            # agregar las columnas nuevas al dataframe original
            cleaned_data = pd.concat([cleaned_data, c_data_interest], axis=1)
        if len(cleaned_data.columns) == 0:
            cleaned_data = c_data_interest

        # Drop de valores duplicados
        if drop_duplicates:
            duplicated_data = cleaned_data.duplicated().sum()
            cleaned_data.drop_duplicates(inplace=True)
            print('Hay ', duplicated_data, ' datos duplicados')

        # Guardar los datos procesados en formato .csv
        cleaned_data_path = os.path.join(clean_data_folder, 'data.csv')
        cleaned_data.to_csv(cleaned_data_path, index=False)

        self.data = cleaned_data
        
    
    def scale(self, scalers=None):

        scale_folder = os.path.join(self.output_folder, 'scale')
        if not os.path.exists(scale_folder):
            os.mkdir(scale_folder)

        data_to_scale = self.data.copy()
        
        if 'anomalies' in data_to_scale.columns:
            data_to_scale.drop('anomalies', axis=1, inplace=True)

        # Obtener los nombres de las columnas num√©ricas y categ√≥ricas
        columnas_numericas = data_to_scale.select_dtypes(include=['number']).columns.tolist()
        columnas_categoricas = data_to_scale.select_dtypes(exclude=['number']).columns.tolist()


        # pipeline numerica
        num_pipe = Pipeline(
            steps=[
                ("log_scaler", FunctionTransformer(func=np.log1p, validate=False)),
                ("minmax_scaler", MinMaxScaler())
            ]
        )

        #pipeline categorica
        cat_steps = []

        if scalers:
            for name, scaler in scalers:
                cat_steps.append((name, scaler))
        
        if cat_steps:
            cat_pipe = Pipeline(cat_steps)
        else:
            cat_pipe = "passthrough"

        # column transformer
        prep = ColumnTransformer(
            transformers=[
                ('num', num_pipe, columnas_numericas),
                ('cat', cat_pipe, columnas_categoricas)
            ]
        )

        data_transformed = prep.fit_transform(data_to_scale)
        data_transformed_df = pd.DataFrame(data_transformed, columns=columnas_numericas + columnas_categoricas)

        # Guardar los datos procesados en formato .csv
        scaled_path = os.path.join(scale_folder, 'scaled_features.csv')
        data_transformed_df.to_csv(scaled_path, index=False)
        data_transformed_df[columnas_numericas] = data_transformed_df[columnas_numericas].astype(float)


        return(data_transformed_df)

    def make_clusters(self, algorithm = 'KMeans',separator=[','], scalers=None, min_clusters=1, max_clusters=10):
        
        if algorithm != 'KMeans':
            raise ValueError("Algoritmo de clustering no v√°lido")
        
        cluster_folder = os.path.join(self.output_folder, 'cluster')
        if not os.path.exists(cluster_folder):
            os.mkdir(cluster_folder)

        # preprocesamiento de datos
        self.clean_data(separator = separator)
        scaled_data = self.scale(scalers=scalers)

        print("Procesamiento de datos completado.")

        # seleccion de columnas numericas
        numeric_columns = scaled_data.select_dtypes(include=['number']).columns

        # Realizar un estudio del codo para determinar el n√∫mero √≥ptimo de clusters
        inertia = []
        max_clusters = np.min([max_clusters, len(scaled_data)])
        for i in range(min_clusters, max_clusters):
            kmeans = KMeans(n_clusters=i, random_state=0)
            kmeans.fit(scaled_data[numeric_columns])
            inertia.append(kmeans.inertia_)

        # Encuentra el "codo" en la curva de la inercia
        diff = np.diff(inertia)
        diff_r = diff[1:] / diff[:-1]
        n_clusters = np.where(diff_r < 0.5)[0][0] + 2  # Escoge el primer punto donde la relaci√≥n sea menor que 0.5

        # Graficar el codo
        plt.figure(figsize=(8, 6))
        plt.plot(range(min_clusters, max_clusters), inertia, marker='o', linestyle='--')
        plt.xlabel('N√∫mero de Cl√∫steres')
        plt.ylabel('Inercia')
        plt.title('M√©todo del Codo')
        plt.savefig(os.path.join(cluster_folder, 'elbow_plot.png'))
        plt.show()
        plt.close()

        print(f"El n√∫mero √≥ptimo de cl√∫steres es: {n_clusters}")

        # pipeline 
        pipeline = Pipeline([
            ('kmeans', KMeans(n_clusters=n_clusters)),  # Aplicar K-Means con n_clusters
            ('pca', PCA(n_components=2))  # Reducir la dimensionalidad a 2 componentes principales
        ])

        pipeline.fit(scaled_data[numeric_columns])
        # Obtener las etiquetas de cl√∫ster de K-Means
        kmeans_labels = pipeline.named_steps['kmeans'].labels_

        # Ajustar el pipeline a los datos
        data_pca = pipeline.fit_transform(scaled_data[numeric_columns])

        # Graficar los datos con colores basados en las etiquetas de cl√∫ster
        plt.scatter(data_pca[:, 0], data_pca[:, 1], c=kmeans_labels, cmap='viridis')
        plt.title('Data PCA con Colores de K-Means Labels')
        plt.xlabel('Dimension 1')
        plt.ylabel('Dimension 2')
        plt.colorbar(label='Etiquetas de Cl√∫ster K-Means')
        plt.savefig(os.path.join(cluster_folder, 'cluster_projection.png'))
        plt.show()
        plt.close()

        self.data['cluster'] = kmeans_labels

        # Guardar los datos con su respectivo cluster en formato .csv
        cluster_data_path = os.path.join(cluster_folder, 'data_clusters.csv')
        self.data.to_csv(cluster_data_path, index=False)

        print("Clustering completado y resultados guardados.")

    def detect_anomalies(self, anomaly_detector=IsolationForest, separator=[','], scalers=None, **kwargs):

        anomalies_folder = os.path.join(self.output_folder, 'anomalies')
        if not os.path.exists(anomalies_folder):
            os.mkdir(anomalies_folder)
        
        # preprocesamiento de datos
        self.clean_data(separator = separator)
        scaled_data = self.scale(scalers=scalers)

        print("Procesamiento de datos completado.")

        # seleccion de columnas numericas
        numeric_columns = scaled_data.select_dtypes(include=['number']).columns

        try:
            detector = anomaly_detector(**kwargs)  # Crear una instancia del detector de anomal√≠as
        except Exception as e:
            raise ValueError(f"No se pudo crear el detector de anomal√≠as: {e}")

        # Entrenar el detector de anomal√≠as
        anomaly_labels = detector.fit_predict(scaled_data[numeric_columns])

        # Aplicar PCA
        pca = PCA(n_components=2)
        data_pca = pca.fit_transform(scaled_data[numeric_columns])

        # Graficar los datos con colores basados en las etiquetas de anomalias
        plt.scatter(data_pca[:, 0], data_pca[:, 1], c=anomaly_labels, cmap='viridis')
        plt.title('Data PCA con Colores de Anomaly Labels')
        plt.xlabel('Dimension 1')
        plt.ylabel('Dimension 2')
        plt.savefig(os.path.join(anomalies_folder, 'anomalies_projection.png'))
        plt.show()
        plt.close()

        self.data['anomalies'] = anomaly_labels

        # Guardar los datos con su respectivo cluster en formato .csv
        anomalies_data_path = os.path.join(anomalies_folder, 'data_anomalies.csv')
        self.data.to_csv(anomalies_data_path, index=False)

        print("Detecci√≥n de anomalias completado y resultados guardados.")
    
    def profile(self):
        self.summarize()
        self.plot_vars()
        self.clean_data()
        self.scale()
        self.make_clusters()
        self.detect_anomalies()
    
    def clearGarbage(self):

        carpeta_a_limpiar = self.output_folder

       # Eliminar archivos en la carpeta
        archivos_en_carpeta = os.listdir(carpeta_a_limpiar)

        for archivo in archivos_en_carpeta:
            ruta_completa = os.path.join(carpeta_a_limpiar, archivo)
            try:
                if os.path.isfile(ruta_completa):
                    os.remove(ruta_completa)
                    print(f"Archivo eliminado: {ruta_completa}")
                elif os.path.isdir(ruta_completa):
                    shutil.rmtree(ruta_completa)
                    print(f"Carpeta eliminada: {ruta_completa}")
            except Exception as e:
                print(f"No se pudo eliminar {ruta_completa}: {e}")

        print(f"Contenido de la carpeta '{carpeta_a_limpiar}' eliminado exitosamente.")
```
